```python
import os
import numpy as np
from PIL import Image
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.models import load_model
import tensorflow.keras.losses as losses
```

```python

def convert_to_black_and_white_folder(input_folder, output_folder, threshold=128):
    os.makedirs(output_folder, exist_ok=True)

    for filename in os.listdir(input_folder):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif', '.tiff')):
            input_path = os.path.join(input_folder, filename)

            try:
                with Image.open(input_path) as img:
                    # KONWERSJA do skali szaroci przed punktowym przetwarzaniem
                    gray_img = img.convert('L')

                    # Prg  binarna mapa czerni i bieli
                    bw_img = gray_img.point(lambda x: 255 if x > threshold else 0, mode='1')

                    # Nazwa pliku
                    name, ext = os.path.splitext(filename)
                    new_filename = f"{name}_czb{ext}"
                    output_path = os.path.join(output_folder, new_filename)

                    bw_img.save(output_path)
                    print(f"Zapisano: {new_filename}")
            except Exception as e:
                print(f"Bd przy pliku {filename}: {e}")

# Przykadowe cieki \ sample source
convert_to_black_and_white_folder(
    input_folder=r"D:\\zdjcia testowe\\kwiaty_same_model_i_data\\kwiaty_same_zdjcia\\flowers",
    output_folder=r"D:\\zdjcia testowe\\kwiaty_same_model_i_data\\kwiaty_same_zdjcia\\flowers_black"
)
```

```python

def convert_to_grey_folder(input_folder,output_folder):
    os.makedirs(output_folder,exist_ok=True)
    
    for filename in os.listdir(input_folder):
        if filename.lower().endswith(('.jpg','.png','.jpeg','.bmp','.gif','.tiff')):
            input_path = os.path.join(input_folder,filename)
            
            try:
                with Image.open(input_path) as img:
                    gray_img=img.convert('L')
                    name, ext=os.path.splitext(filename)
                    new_filename=f"{name}_gray{ext}"
                    output_path=os.path.join(output_folder,new_filename)
                    gray_img.save(output_path)
            except Exception as e:
                print(f"Bd przy {filename}: {e}")

#przykadowe cieki / sample source
convert_to_grey_folder(
    input_folder=r"D:\\zdjcia testowe\\kwiaty_same_model_i_data\\kwiaty_same_zdjcia\\flowers",
    output_folder=r"D:\\zdjcia testowe\\kwiaty_same_model_i_data\\kwiaty_same_zdjcia\\flowers_gray"
)
```
